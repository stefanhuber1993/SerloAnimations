```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import numpy as np

x = np.linspace(-2.5,2.5,100)

#matplotlib.rc('font', family='serif', serif='cm10')

#matplotlib.rc('text', usetex=True)
#matplotlib.rcParams['text.latex.preamble'] = [r'\boldmath']

color1 = 'red'
color2 = 'blue'
```

```python
fig,ax=plt.subplots(figsize=(3.5,3.5), dpi=150)

# Deactivate normal axes
for side in ['bottom','right','top','left']:
    ax.spines[side].set_visible(False)
plt.xticks([]) # labels 
plt.yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')

# Make custom axes with arrows
ax.arrow(-3, 0, 6, 0., fc='k', ec='k', lw = 1, 
         head_width=0.1, head_length=0.2,
         length_includes_head= True, clip_on = False) 
ax.arrow(0, -3, 0, 6, fc='k', ec='k', lw = 1, 
         head_width=0.1, head_length=0.2,
         length_includes_head= True, clip_on = False) 

# Plot Untervektorraum
plt.plot(-x, x, c=color1)
plt.text(2.0,2.5,'$V=‚Ñù^2$')
plt.text(1,-0.8,'$U$', color=color1)
plt.savefig('LineareAlgebra1/NebenklassenEinesVektorraums/R2_mit_Untervektorraum.svg')

# Additionally plot Nebenklasse
plt.plot(-x[:-25], x[:-25]+2, c=color2)
plt.text(0.4,1.8,'$v+U$', color=color2)
ax.arrow(0, 0, -1, 3,fc='k', ec='k', lw = 1, 
         head_width=0.1, head_length=0.2,
         length_includes_head= True, clip_on = False) 
plt.text(-1,1.8,'$v$')
plt.savefig('LineareAlgebra1/NebenklassenEinesVektorraums/R2_mit_Untervektorraum_und_Nebenklasse.svg')
```

```python
fig,ax=plt.subplots(figsize=(3.5,3.5), dpi=150)

# Deactivate normal axes
for side in ['bottom','right','top','left']:
    ax.spines[side].set_visible(False)
plt.xticks([]) # labels 
plt.yticks([])
ax.xaxis.set_ticks_position('none') # tick markers
ax.yaxis.set_ticks_position('none')

# Make custom axes with arrows
ax.arrow(-3, 0, 6, 0., fc='k', ec='k', lw = 1, 
         head_width=0.1, head_length=0.2,
         length_includes_head= True, clip_on = False) 
ax.arrow(0, -3, 0, 6, fc='k', ec='k', lw = 1, 
         head_width=0.1, head_length=0.2,
         length_includes_head= True, clip_on = False) 

# Plot Untervektorraum
plt.plot(-x, x, c=color1)
plt.text(2.7,2.5,'$V=‚Ñù^2$')
plt.text(1,-0.8,'$U$', color=color1)

# Additionally plot Nebenklasse
for i in np.linspace(-6,6,100):
    plt.plot(-x, x+i, c=color2, alpha=0.25)
    
plt.xlim(-3,3); plt.ylim(-3,3)

plt.savefig('LineareAlgebra1/NebenklassenEinesVektorraums/R2_mit_Untervektorraum_und_vielen_Nebenklassen.svg')
```
